# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50) # min / max / tam
meses = np.arange(1, 51)
print(meses)
```

```python
plt.plot(meses, vendas)

plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.axis([1, 50, 0, max(vendas)+200])
plt.show()
```

- Editando o Gráfico de Linha para outros tipos


```python
plt.plot(meses, vendas, "r--")

plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.axis([1, 50, 0, max(vendas)+200])
plt.show()
```

```python
plt.plot(meses, vendas, "ro")

plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.axis([1, 50, 0, max(vendas)+200])
plt.show()
```

```python
plt.plot(meses, vendas, "bs")

plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.axis([1, 50, 0, max(vendas)+200])
plt.show()
```

```python
plt.plot(meses, vendas, "g^")

plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.axis([1, 50, 0, max(vendas)+200])
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, "ro", t, t**2, "bs", t, t**3, "g^")

plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.axis([0, 5, 0, max(t**3)+1])
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?

plt.plot(meses, vendas, color="pink", linewidth=5.0)

plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.axis([1, 50, 0, max(vendas)+200])
plt.show()
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.show()
```

- Barras

```python
plt.bar(meses, vendas)
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação -> figura


```python
plt.figure(1, figsize=(12,3)) # primeira figura	(1 é somente um identificador) // figsize é o tamanho da figura

plt.subplot(1, 3, 1) # 1 linha, 3 colunas, primeiro gráfico (de tres)
plt.plot(meses, vendas)

plt.subplot(1, 3, 2) # 1 linha, 3 colunas, segundo gráfico (de tres)
plt.scatter(meses, vendas)

plt.subplot(1, 3, 3) # 1 linha, 3 colunas, terceiro gráfico (de tres)
plt.bar(meses, vendas)

plt.show()
```
